## 직업탐구하기 pt.1

``groupby``문법을 사용해서 평균나이가 어린 순으로 직업을 나열하자

## 데이터 합치기

데이터는 합칠 때 **pandas**를 이용할 수 있다.

* inner join : 교집합만 합치는 것
* left outer join : 왼쪽데이터 프레임에 존재하는 것을 합치는 것
* right outer join : 오른쪽데이터 프레임에 존재하는 것을 합치는 것
* full outer join : 양 쪽에 있는 모든 데이터를 합치는 것


```python
import pandas as pd

price_df = pd.read_csv('data/vegetable_price.csv')
quantity_df = pd.read_csv('data/vegetable_quantity.csv')
```

```python
price_df
```

```python
quantity_df
```

```python
pd.merge(price_df, quantity_df, on = 'Product')  # default 값은 inner Join
```

```python
pd.merge(price_df, quantity_df, on = 'Product', how = 'left')
```

```python
pd.merge(price_df, quantity_df, on = 'Product', how = 'right')
```

```python
pd.merge(price_df, quantity_df, on = 'Product', how = 'outer')
```
