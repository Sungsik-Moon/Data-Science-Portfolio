## DataFrame에 값 추가/삭제

```python
iphone_df.drop('iPhone XR', axis = 'index', inplace = True)
```

## 서류전형 합격 여부

```python
# 서울항공에 입사하기 위해 토익점수가 LC, RC 각각 최소 250점, 그리고 점수 총합은 600이 넘어야한다.

# DataFrame에 '합격여부' 라는 column을 추가하고 합격한 지원자는 True 불합격한 지원자는 False가 입력되도록 코딩하라

import pandas as pd

df = pd.read_csv('data/toeic.csv', index_col = 0)
df
```

```python
df['LC'] | df['RC'] > 249
```

```python
(df['LC'] + df['RC']) > 599
```

```python
df["합격 여부"] = (df['LC'] & df['RC'] > 249) & ((df['LC'] + df['RC']) > 599)
df
```

## index/column 설정하기

```python
import pandas as pd

liverpool_df = pd.read_csv('data/liverpool.csv', index_col = 0)
liverpool_df
```

```python
liverpool_df.rename(columns = {'position':'Position'})
    
#rename은 이름을 바꿔주는 함수
```

``rename``함수가 기존의 DataFrame을 건들지 않고 잠시동안 새로운 DataFrame을 만들어내기 때문에 다시 **liverpool_df**를 출력할 때
position이 변경되지 않은 채 인 것

```python
liverpool_df
```

```python
liverpool_df.rename(columns = {'position':'Position'}, inplace = True)

liverpool_df
```

위 **{'position':'Position'}** 은 파이선 딕셔너리인데 이를 이용해서 여러개의 컬럼을 바꿔줄 수도 있다.

```python
liverpool_df.rename(columns = {'position':'Position', 'born':'Born', 'number':'Number', 'nationality':'Nationality'}, inplace = True)
liverpool_df
```

```python
liverpool_df.index.name = 'Player Name'
```

```python
liverpool_df
```

```python
#만약 등번호를 인덱스로 사용하고 싶다면?

liverpool_df.set_index('Number')
```

위와 같이 실행하면 **Number**가 인덱스명으로 잘 넘어오기는 했는데 선수 이름이 사라져버리게 된다.

이를 위해 기존 인덱스를 새로운 컬럼에 지정해주는 것이 좋다

```python
liverpool_df['Player Name'] = liverpool_df.index
```

```python
liverpool_df
```

```python
liverpool_df.set_index('Number', inplace = True)
```

```python
liverpool_df
```

인덱스는 어떤 row를 받아오기 위해 사용되기 때문에 겹치는 값이 없는 것이 좋다.

때문에 겹칠 가능성이 있는 선수의 이름보다는 중복이 있을 수 없는 선수의 등번호가 인덱스로 활용되기 좋은 것이다.
