## 코드잇 대학교: 수강신청 준비하기

2000명의 코드잇 대학교 학생들이 수강신청을 했습니다.

수강신청에는 다음 3개의 조건이 있습니다.

* "information technology"과목은 심화과목이라 1학년은 수강할 수 없습니다.
* "commerce"과목은 기초과목이고 많은 학생들이 듣는 수업이라 4학년은 수강할 수 없습니다.
* 수강생이 5명이 되지 않으면 강의는 폐강되어 수강할 수 없습니다.

DataFrame에 "status"라는 컬럼을 추가하고 수강이 가능한 상태이면 'allowed' 그렇지 않으면 'not allowed'를 넣어주세요

```python
import pandas as pd

enrol_df = pd.read_csv('data/enrolment_1.csv', index_col = 0)
enrol_df
```

```python
enrol_df['status'] = 'allowed'
enrol_df.loc[(enrol_df['course name'] == 'information technology') & (enrol_df['year'] == 1), 'status'] = 'not allowed'
enrol_df.loc[(enrol_df['course name'] == 'commerce') & (enrol_df['year'] == 4), 'status'] = 'not allowed'
```

```python
enrol_df['status'].value_counts()
```

```python
allowed = enrol_df['status'] == "allowed"  # 상태가 allowed인 신청강좌만 집계한다.
allowed
```

```python
course_counts = enrol_df['course name'].value_counts()# course counts를 시리즈로 저장
```

```python
course_counts
```

```python
type(course_counts)
```

```python
closed_courses = list(course_counts[course_counts < 5].index)  # course counts가 5 이하인 인덱스의 리스트
```

```python
closed_courses
```

```python
len(closed_courses)  # 총 214개의 강좌가 인원 수 미달로 확인되었다.
```

```python
type(closed_courses)
```

```python
for i in range(len(enrol_df)):
    if enrol_df.loc[i, 'course name'] in closed_courses:
        enrol_df.loc[i, 'status'] = 'not allowed'
```

## 코드잇 대학교 수강신청  정답지

```python
import pandas as pd

df = pd.read_csv('data/enrolment_1.csv')

df['status'] = 'allowed'
```

일단 status 값에 해당사항이 없다면 모두 허용되는 과목이므로 status 컬럼을 만들어주고 'allowed'로 채워준다

그리고나서 조건 1 ' information technology'과목을 1학년이 신청한 경우에 관한 조건문을 만들어준다

```python
boolean1 = df["course name"] == "information technology"
boolean2 = df["year"] == 1
```

```python
df.loc[boolean1 & boolean2, 'status'] = 'not allowed'
```

boolean1, 2를 통해 인덱싱된 status 컬럼에 'not allowed'를 씌워줬다.

다음은 commerce 과목을 이수하려는 4학년 학생들의 수강신청 결과를 'not allowed'로 변경해주자

```python
boolean3 = df['course name'] == 'commerce'
boolean4 = df['year'] == 4
```

```python
df.loc[boolean3 & boolean4 , 'status'] = 'not allowed'
```

전에 했던 것과 같은 과정으로 해당하는 학생들의 status 를 not allowed로 변경해주자
